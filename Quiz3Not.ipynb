```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def Harrisp(img):
    gray = np.float32(img)
    # blockSize=2, aperture size sobel = 3, k =0.04
    dst = cv2.cornerHarris(gray,2,3,0.04)
    # Threshold for an optimal value, it may vary depending on the image.
    
    gray[dst>0.01*dst.max()]=[20]
    return gray
```

```python
def Harris(img):
    
    img = np.float32(img)
    
    Sx = np.array([[-1, 0, 1], [-2, 0, 2],[-1, 0, 1]])
    Sy = np.array([[-1, -2, -1], [0, 0, 0],[1, 2, 1]])
    
    Ix = cv2.filter2D(img, -1, Sx)
    Iy = cv2.filter2D(img, -1, Sy)
    
    # Ix = cv2.Sobel(imgg,cv2.CV_64F,1,0,ksize=3)  # x
    # Iy = cv2.Sobel(imgg,cv2.CV_64F,0,1,ksize=3)  # y
    
    Ixx = Ix*Ix
    Iyy = Iy*Iy
    Ixy = Ix*Iy
    
    Sxx = cv2.GaussianBlur(Ixx,(5,5), cv2.BORDER_DEFAULT)
    Syy = cv2.GaussianBlur(Iyy,(5,5), cv2.BORDER_DEFAULT)
    Sxy = cv2.GaussianBlur(Ixy,(5,5), cv2.BORDER_DEFAULT)
    
    
    k = 0.04
    
    R = (Sxx*Syy - Sxy*Sxy) - k*((Sxx + Syy)**2)
    
    img[R>0.09*R.max()]=[20]
    
    return img
```

```python
def noyse(img, mean, std):
    
    img = np.float32(img)   
    img = img + np.random.normal(mean, std, imgs.shape)
    # img = img/(img.max()/255)
    
    return img    
```

```python
row = 200
cols = 300
   
imgs = np.zeros((row,cols))
imgs[:,:] = 255

imgs[:,np.arange(int(cols/5),cols,int(cols/5))] = 0
imgs[np.arange(int(row/5),row,int(row/5)),:] = 0

plt.figure(6)
plt.imshow(imgs, cmap = 'gray')

t = 1000
std = 0 

imgf = np.zeros((row, cols, t))
imgfp = np.zeros((row, cols, t))

resul = np.zeros((7,t))
for i in np.arange(0,t):
    mean = 0
    resul[0,i] = std
    img = noyse(imgs, mean, std)
    imgf[:,:,i] = Harris(img) 
    # plt.imsave('./plot/image_' + str(i) + '.jpg', np.uint8(imgf[:,:,i]), cmap='gray')
    imgfp[:,:,i] = Harrisp(img) 
    # plt.imsave('./plot/imagep_' + str(i) + '.jpg', np.uint8(imgfp[:,:,i]), cmap='gray')
    std = std + 0.05


for k in np.arange(0,t): 
    contc = 0
    conte = 0
    contm = 0
    contmp = 0
    contcp = 0
    contep = 0
    
    for i in np.arange(0,row):
        for j in np.arange(0,cols):
    
            if (imgf[i,j,k] == 20 and imgf[i,j,0] == 20):
                contc = contc + 1
            if (imgf[i,j,k] != 20 and imgf[i,j,0] == 20):
                contm = contm + 1
            if (imgf[i,j,k] == 20 and imgf[i,j,0] != 20):
                conte = conte + 1
            
            
            if (imgfp[i,j,k] == 20 and imgfp[i,j,0] == 20):
                contcp = contcp + 1        
            if (imgfp[i,j,k] != 20 and imgfp[i,j,0] == 20):
                contm = contm + 1
            if (imgfp[i,j,k] == 20 and imgfp[i,j,0] != 20):
                contep = contep + 1
    
    resul[1,k] = contc
    resul[2,k] = conte
    resul[3,k] = contm
    resul[4,k] = contcp
    resul[5,k] = contep
    resul[6,k] = contmp
n = row*cols
rstd = resul[0,:]
rms1 = np.sqrt(((resul[0,:] - resul[0,0])**2)/n)
rms2 = np.sqrt(((resul[1,:] - resul[1,0])**2)/n)
rms3 = np.sqrt(((resul[2,:] - resul[2,0])**2)/n) 
rmsp1 = np.sqrt(((resul[3,:] - resul[3,0])**2)/n)
rmsp2 = np.sqrt(((resul[4,:] - resul[4,0])**2)/n) 
rmsp3 = np.sqrt(((resul[5,:] - resul[5,0])**2)/n) 
```

```python
plt.figure(0)
plt.plot(rstd, rms1)
plt.title('Harris Implementação Acertos')
plt.figure(1)
plt.plot(rstd,rms2)
plt.title('Harris Implementação Perda')
plt.figure(2)
plt.plot(rstd,rms3)
plt.title('Harris Implementação Erro')

plt.figure(3)
plt.title('Harris Opencv Acertos')
plt.plot(rstd,rmsp1)
plt.figure(4)
plt.title('Harris Opencv Perda')
plt.plot(rstd,rmsp2)
plt.figure(5)
plt.title('Harris Opencv Erro')
plt.plot(rstd,rmsp3)
```
